```python
# Basic Inheritance Example

# Parent Class (Superclass)
class Electronics:
    def __init__(self):
        self.power_status="OFF"
    
    def turn_on(self):
        self.power_status="ON"
        print("Device is now powered ON\n")

# Child Class (Subclass)
class Drone(Electronics):
    def fly(self):
        if self.power_status=="ON":
            print("The drone is now flying")
        else:
            print("Cannot fly:Power is OFF")

# Usage
my_drone=Drone()
my_drone.turn_on()
my_drone.fly()

```

```python
# Encapsulation

#Capsule of data + realted functions in a single unit

class Account:
    def __init__(self,bal,acc):
        self.balance=bal
        self.account=acc
    def debit(self,amount):
        self.balance-=amount
        print(f"Rs. {amount} was debited ")
    def credit(self,amount):
        self.balance+=amount
        print(f"Rs. {amount} was credited\nBal ",self.get_balance())
    
    def get_balance(self):
        return self.balance



acc=Account(10,1212)
print(acc.balance)
print(acc.account)
acc.debit(1)
acc.credit(500)

```
