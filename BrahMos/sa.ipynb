```python
import pandas as pd
import string

df = pd.DataFrame({'numbers': [12, 13, 14, 15, 16, 17, 18, 19, 20]})

alphabet = string.ascii_lowercase

df['letters'] = df['numbers'].apply(lambda x: alphabet[x-1])
```

```python
df
```

```python

filtered_df = df[~df['letters'].str.contains('e', na=False)]
filtered_df
```

```python

result = df.groupby('letters')['letters'].sum()
result
```

```python
# Goutham question 
import pandas as pd
data = {'customer_id': [1, 1, 2, 3],
        'amount': [100, 200, 50, 500]}
df = pd.DataFrame(data)
customer_spend = df.groupby('customer_id')['amount'].sum().reset_index()
def segment_customer(total_spend):
    if total_spend > 300:
        return 'High Value'
    elif 100 <= total_spend <= 300:
        return 'Medium'
    else:
        return 'Low'
customer_spend['Segment'] = customer_spend['amount'].apply(segment_customer)
print(customer_spend)
```

```python
#Kowshik question
import pandas as pd

l1 = list(range(1, 101))
l2 = [l1[i:i+10] for i in range(0, 100, 10)]

l3 = [list(range(i, i + 10, 2)) for i in range(2, 101, 10)]

df = pd.DataFrame(l2)
df.iloc[:, 0] = ['A', 'A', 'B', 'C', 'D', 'E', 'E', 'E', 'F', 'F']
df_filtered = df[df.iloc[:, 0] < 'E']
grouped_sum = df.groupby(df.columns[0])[df.columns[3]].sum()
print("1:", l1)
print("2:", l2)
print("3:", l3)
print("4:\n", df)
print("5:\n", df_filtered)
print("6:\n", grouped_sum)



```
